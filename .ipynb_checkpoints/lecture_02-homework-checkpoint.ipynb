## Homework 1 - Bollinger Bands

A **Bollinger Band®** is a technical analysis tool defined by a set of **trendlines plotted two standard deviations (positively and negatively) away from a simple moving average (SMA)** of a security's price, but which can be adjusted to user preferences. Bollinger Bands® were developed and copyrighted by the famous technical trader John Bollinger, designed to  identify when an asset is oversold or overbought. Many traders believe the closer the prices move to the upper band, the more overbought the market is, and the closer the prices move to the lower band, the more oversold the market is <a href="https://www.investopedia.com/terms/b/bollingerbands.asp">[1]</a>.

**Task 1** consists in reading the whole APPL (Apple Inc.) dataset available in the _data_ directory, computing and plotting the Bollinger bands of the adjusted close price, moving average (20 days window size) and the adjusted close price itself. **Task 2** consists in identifying potential buying and selling opportunities and plotting vertical lines in the plot: green corresponding to buy signals, red to sell signals.

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.options.display.width = 1200

import matplotlib
matplotlib.rc('figure', figsize=(10,3), dpi=200)

ALL_SYMBOLS = ["AAPL"]
DATE_RANGES = ('2019-09-01', '2021-09-01')
```

```python
def adjusted_close_symbols(symbols, date_range):
    """ Joins an empty DataFrame with index, with multiple stocks dataframes. """

    # dates is formatted as YYYY-MM-DD
    dates = pd.date_range(*date_range)

    # An empty dataframe
    df1 = pd.DataFrame(index=dates) # the dataframe only with the index

    for sym in symbols:
        path_to_file = os.path.join("data", "%s.csv" % sym)
        dfSYM = pd.read_csv(path_to_file, index_col="Date", parse_dates=True, usecols=["Date", "Adj Close"]) # na_values=['nan']
        dfSYM = dfSYM.rename(columns= {'Adj Close': sym})
        df1 = df1.join(dfSYM, how="inner")
    return df1
```

```python
df = adjusted_close_symbols(ALL_SYMBOLS, DATE_RANGES)
print(df)
```

```python
# plotting procedure 
ax = df.plot(title="Adjusted Closes", color="g")
ax.set_xlabel("Date")
ax.set_ylabel("Price")
plt.legend(loc='upper left') # better visualize the legend
plt.show()
```

```python
def fireSignal(indicators, toCooldown=1):
    """
    Given an array indicating if at a specific time the "adj close" was above/under the UB/LB,
    return the indixes of the crossing.
    We continue segnalling until we reach 'cooldown' steps (time delta). At that point, another crossing must occurr
        before starting segnalling again
    """
    cd = toCooldown

    # replace with Nan
    s_nan = indicators.where(indicators!=0)
    isnaIndexes = s_nan.isna() # where to add the +1 after the cumsum
    
    # apply cum sum - ignores the nan
    s_nan = s_nan.cumsum()
    
    s_refilled = s_nan.fillna(method="ffill")
    s_refilled[isnaIndexes] += 1
    s_shift = s_refilled.shift(cd)    
    tmp = s_refilled - s_shift
    tmp.fillna(cd-1) # fill the initial shifted position with a value that will return a signal (we are within the cooldown timeDelta)
    tmp[isnaIndexes] = cd # switch off the signal for those position whose original signal was 'off'    
    
    # Apply the linear separation between valid/invalid signal
    tmp = np.where(tmp>=cd, 0,1)
    
    
    return tmp.astype(bool)
```

## "1TimeStep" startCooldown version

```python
wCooldown= True
window_size = 20 # days
F = 2
df["AAPL_rm({})".format(window_size)] = df.rolling(window_size).mean()  # add a column to the dataframe
df["ub"] = df["AAPL_rm({})".format(window_size)] + (df["AAPL"].rolling(window_size).std() * F)
df["lb"] = df["AAPL_rm({})".format(window_size)] - (df["AAPL"].rolling(window_size).std() * F)

isAboveUb = df["AAPL"]>=df["ub"]
isBelowLb = df["AAPL"]<=df["lb"]


ax = df.plot(linewidth=1)
if wCooldown:
    # for the sell signals
    t_shift = isAboveUb.shift(+1).astype("bool")
    t_shift[0] = isAboveUb[0]
    sellSignals = isAboveUb & -t_shift
    sellSignals[0] = isAboveUb[0]
    ax.vlines(x=df[sellSignals].index.values, ymin=df["lb"].min(), ymax=df["ub"].max(), color='r', linestyle='--', lw=0.5, label="sell")
    # for the buy signals
    t_shift = isBelowLb.shift(+1).astype("bool")
    t_shift[0] = isBelowLb[0]
    buySignals = isBelowLb & -t_shift
    buySignals[0] = isBelowLb[0]
    ax.vlines(x=df[buySignals].index.values, ymin=df["lb"].min(), ymax=df["ub"].max(), color='g', linestyle='--', lw=0.5, label="buy")
else: # No Cooldown here
    ax.vlines(x=df[isAboveUb].index.values, ymin=df["lb"].min(), ymax=df["ub"].max(), color='r', linestyle='--', lw=0.5, label="sell")
    ax.vlines(x=df[isBelowLb].index.values, ymin=df["lb"].min(), ymax=df["ub"].max(), color='g', linestyle='--', lw=0.5, label="buy")

plt.legend(fontsize=8)
plt.show()
```

## Variable "TimeStep" startCooldown

```python
wCooldown= True
tc = 1 # fires the signal for `tc` steps before starting the cooldown

window_size = 20 # days
F = 2
df["AAPL_rm({})".format(window_size)] = df.rolling(window_size).mean()  # add a column to the dataframe
df["ub"] = df["AAPL_rm({})".format(window_size)] + (df["AAPL"].rolling(window_size).std() * F)
df["lb"] = df["AAPL_rm({})".format(window_size)] - (df["AAPL"].rolling(window_size).std() * F)

isAboveUb = df["AAPL"]>=df["ub"]
isBelowLb = df["AAPL"]<=df["lb"]


ax = df.plot(linewidth=1)
if wCooldown:
    sellSignals = fireSignal(isAboveUb, toCooldown=tc)
    ax.vlines(x=df[sellSignals].index.values, ymin=df["lb"].min(), ymax=df["ub"].max(), color='r', linestyle='--', lw=0.5, label="sell")
    buySignals = fireSignal(isBelowLb, toCooldown=tc)
    ax.vlines(x=df[buySignals].index.values, ymin=df["lb"].min(), ymax=df["ub"].max(), color='g', linestyle='--', lw=0.5, label="buy")

    
else: # No Cooldown here
    ax.vlines(x=df[isAboveUb].index.values, ymin=df["lb"].min(), ymax=df["ub"].max(), color='r', linestyle='--', lw=0.5, label="sell")
    ax.vlines(x=df[isBelowLb].index.values, ymin=df["lb"].min(), ymax=df["ub"].max(), color='g', linestyle='--', lw=0.5, label="buy")

plt.legend(fontsize=8)
plt.show()
```
